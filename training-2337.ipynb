```python
def nor(x):
    li = []
    std = x.std()
    mean = x.mean()
    for i in range(0,len(x)):
        a = float(x.iloc[i])
        li.append((a-mean)/std)
    li = pd.Series(li).as_matrix()
    return li
```

```python
time_steps=17
X_batch1 =[]
Y_batch1 = []
ll=[]
ll1=[]
for i in range(X_batch.shape[0]-time_steps):
    try:
        ll.append(X_batch[i:i+time_steps])
        ll1.append(Y_batch[i:i+time_steps])
    except:
        pass
    
    X_batch1.append(ll)
    ll=[]
    Y_batch1.append(ll1)
    ll1 = []
        
```

```python
start = 0
BATCH_SIZE=2
TIME_STEPS = 17
BATCH_N = int(X_batch.shape[0]/BATCH_SIZE/TIME_STEPS)
INPUT_SIZE = 6
CELL_SIZE = 3
OUTPUT_SIZE=27
```

```python
X_batch1 = np.array(X_batch1)
Y_batch1 = np.array(Y_batch1)
```

```python
X_batch1  =X_batch1.reshape(-1,17,6)
Y_batch1 = Y_batch1.reshape(-1,17,1)
```

```python
X_batch2=[]
Y_batch2=[]
```

```python
for i in range(0,1140,2):
    llll=[]
    llll1=[]
    llll.append(X_batch1[i])
    llll.append(X_batch1[i+1])
    llll1.append(Y_batch1[i])
    llll1.append(Y_batch1[i+1])
    X_batch2.append(llll)
    Y_batch2.append(llll1)
```

```python
X_batch=np.array(X_batch2).reshape(570,BATCH_SIZE,TIME_STEPS,INPUT_SIZE)
```

```python
Y_batch = np.array(Y_batch2).reshape(570,BATCH_SIZE,TIME_STEPS,1)
```

```python
LR = 0.00001
```

```python
model =Sequential()
model.add(LSTM(
    batch_input_shape=(BATCH_SIZE,TIME_STEPS, INPUT_SIZE),
    output_dim=32,
    return_sequences=True,
    stateful=True,
))
model.add(Dropout(0.2))
model.add(LSTM(
    output_dim=16,
    return_sequences=True,
    stateful=True,
))
model.add(Dropout(0.2))
model.add(TimeDistributed(Dense(OUTPUT_SIZE)))
model.add(Dense(8,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
adam = Adam(LR)
model.compile(optimizer=adam,
              loss='binary_crossentropy',)
# model.load_weights('3029withpercentage.h5', by_name=True)
```

```python
cl = []
```

```python
epoch = 3001
for epoch in range(epoch):
    # data shape = (batch_num, steps, inputs/outputs)
    for i in range(0,BATCH_N):
        X = X_batch[i]
        Y = Y_batch[i]
        cost = model.train_on_batch(X,Y)
        pred = model.predict(X, BATCH_SIZE)
        cl.append(cost)
    if epoch % 100  == 0:
        print('cl',np.average(cl))
#         l = model.layers[0].get_weights()
        print('train cost: ', cost,'epoch: ',epoch)
        cl=[]
```
